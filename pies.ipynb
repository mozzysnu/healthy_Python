```
#엑셀 데이터 업로드
from google.colab import files
uploaded = files.upload()
```

```
#데이터 pandas로 표시
import io
import numpy as np
import pandas as pd
df = pd.read_excel(io.BytesIO(uploaded['data.xlsx']))
df.pop("n")
df
```

```
#pandas데이터 dictionary형태로 바꾸기
df.set_index('지역구').T.to_dict('list')
piedata1 = df.set_index('지역구').T.to_dict('list')
piedata1
```

```
#matplot으로 pie graph만들기
from pandas import DataFrame
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

plt.rc('font', family='NanumBarunGothic') 

region_list = list(piedata1.keys())
for region in region_list:
    dt1 = DataFrame(piedata1, columns=[region])
    my_labels = 'dentistry', 'dermatology',	'ophthalmology', 'plastic surgery', 'internal medicine', 'pediatric hospital', 'orthopedics', 'psychiatry'
    my_colors = ['lightcoral', 'peru', 'khaki', 'yellowgreen', 'c', 'cadetblue', 'slategrey', 'rebeccapurple']
    my_explode = (0.1, 0, 0, 0, 0, 0, 0, 0)
    plt.pie(dt1, labels=my_labels, autopct='%1.1f%%', startangle=15, shadow = True, colors=my_colors, explode=my_explode)
    plt.title(region)
    plt.axis('equal')
    plt.savefig(str(region))
    plt.show()
    files.download(str(region)+'.png') 
```
